```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1:]
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
pca = PCA(2)
X_pca = pca.fit_transform(X_scaled)
X_pca = pd.DataFrame(X_pca)
X_pca
```

```python
dict = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
y = df['Species'].map(dict)
plt.scatter(X_pca.iloc[:, 0], X_pca.iloc[:,1], c=y, cmap='tab10')
plt.xlabel("PCA_Component1")
plt.ylabel("PCA_Component2")
plt.title("Reduced 2D Data")
```

```python
X_pca['Species'] = df.iloc[:,-1]
X_pca
```

```python
X_pca.to_csv("iris_dataset_2D.csv", index=False)
print("File created successfully!!!")
```
