```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn . linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
train_df = pd.read_csv("Train.csv")
valid_df = pd.read_csv("Validation.csv")
test_df = pd.read_csv("Test.csv")
```

```python
X_train = train_df.iloc[:,0:1]
y_train = train_df.iloc[:,1]

X_valid = valid_df.iloc[:,0:1]
y_valid = valid_df.iloc[:,1]

X_test = test_df.iloc[:,0:1]
y_test = test_df.iloc[:,1]
```

### Simple Linear Regression

```python
regressor = LinearRegression ()

regressor.fit(X_train, y_train)


y_train_pred = regressor.predict(X_train)
y_valid_pred = regressor.predict(X_valid)
y_test_pred = regressor.predict(X_test)
```

```python
# Scatter Plot and Regressio Line of Training Data

plt.scatter(X_train, y_train, label="Actual values")
plt.plot(X_train, y_train_pred, label="Predicted values", color="red")
plt.legend()
plt.xlabel('PT08.S1(CO)')
plt.ylabel('CO(GT)')
plt.show()
```

### RMSE Computation

```python
RMSE_train = math.sqrt(mean_squared_error(y_train,y_train_pred))
RMSE_val = math.sqrt(mean_squared_error(y_valid,y_valid_pred))
RMSE_test = math.sqrt(mean_squared_error(y_test,y_test_pred))

print(f'RMSE values \n{RMSE_train}\n{RMSE_val}\n{RMSE_test}')
```

```python
# Scatter Plot

plt.scatter(y_test_pred, y_test)
plt.title("Actual v/s Predicted Values of 'CO Concentration' for Test Data")
plt.xlabel("Actual Test Values")
plt.ylabel("Predicted Test Values")
```
