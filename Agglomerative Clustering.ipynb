```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = pd.read_csv("iris_dataset_2D.csv")
X, y = df.iloc[:, :-1], df.iloc[:, -1]
X
```

### Agglomerative Clustering

#### Ward Linkage

```python
agnes = AgglomerativeClustering(n_clusters=3, metric='euclidean', memory=None, connectivity=None, compute_full_tree='auto', linkage='ward', distance_threshold=None, compute_distances=False).fit(X)
cluster_labels = agnes.labels_

plt.scatter(x=X.iloc[:,0],y=X.iloc[:,1], c = cluster_labels, cmap='tab10', alpha=0.7)
plt.title('Visualization of Clusters (linkage=Ward)')
plt.xlabel('PCA component_1')
plt.ylabel('PCA component_2')
plt.show()

total = 0

for cluster_id in range(agnes.n_clusters_):
    cluster_indices = np.where(cluster_id == cluster_labels)[0]
    true_labels = y[cluster_indices]

    most_freq = true_labels.mode()[0]
    freq = (most_freq == true_labels).sum()
    print(f'Purity Score for cluster {cluster_id}: {(float(freq/len(true_labels))*100):.4f}')

    total += freq 

print(f'\nOverall Purity Score: {(float(total/len(X))*100):.4f}')
```

#### Best Linkage Method is 'ward' with a purity score of 80.67

```python
# linkage_methods = ['ward', 'complete', 'average', 'single']
linkage_methods=['ward']
threshold=6
best_purity = float('-inf')
for i in range(1,threshold+1):
    for link in linkage_methods:
        agnes = AgglomerativeClustering(n_clusters=i, metric='euclidean', memory=None, connectivity=None, compute_full_tree='auto', linkage=link, distance_threshold=None, compute_distances=False).fit(X)
        cluster_labels = agnes.labels_
        
        plt.scatter(x=X.iloc[:,0],y=X.iloc[:,1], c = cluster_labels, cmap='tab10', alpha=0.7)
        plt.title(f'Visualization of Clusters (linkage={link})')
        plt.xlabel('PCA component_1')
        plt.ylabel('PCA component_2')
        plt.show()
        
        total = 0
        
        for cluster_id in range(agnes.n_clusters_):
            cluster_indices = np.where(cluster_id == cluster_labels)[0]
            true_labels = y[cluster_indices]
        
            most_freq = true_labels.mode()[0]
            freq = (most_freq == true_labels).sum()
            print(f'Purity Score for cluster {cluster_id}: {(float(freq/len(true_labels))*100):.4f}')
        
            total += freq 
        purity_score = (float(total/len(X))*100)
        if (purity_score > best_purity):
            best_linkage = link
            best_purity = purity_score
            optimal_clusters = i

        linked = linkage(X, method = link)
        dendrogram(linked)
    
        plt.title(f'Dendrogram for AGNES (linkage method : {link})')
        plt.axhline(y=6,color='red',linestyle='dashed',label='threshold')
        plt.show()
        
        print(f'\nOverall Purity Score: {purity_score:.4f}')
        print("-----------------------------------------------------------------------------------------------------------------------")
```

```python
print(f'Best linkage method is {best_linkage}, with {optimal_clusters} as optimal clusters yields purity score {best_purity}')
```

```python
agnes = AgglomerativeClustering(n_clusters=optimal_clusters, metric='euclidean', memory=None, connectivity=None, compute_full_tree='auto', linkage=best_linkage, distance_threshold=None, compute_distances=False).fit(X)
cluster_labels = agnes.labels_

plt.scatter(x=X.iloc[:,0],y=X.iloc[:,1], c = cluster_labels, cmap='tab10', alpha=0.7)
plt.title(f'Visualization of Clusters (linkage={best_linkage})')
plt.xlabel('PCA component_1')
plt.ylabel('PCA component_2')
plt.show()

total = 0

for cluster_id in range(agnes.n_clusters_):
    cluster_indices = np.where(cluster_id == cluster_labels)[0]
    true_labels = y[cluster_indices]

    most_freq = true_labels.mode()[0]
    freq = (most_freq == true_labels).sum()
    print(f'Purity Score for cluster {cluster_id}: {(float(freq/len(true_labels))*100):.4f}')

    total += freq 

linked = linkage(X, method = best_linkage)
dendrogram(linked)
    
plt.title(f'Dendrogram for AGNES (linkage method : {best_linkage})')
plt.axhline(y=6,color='red',linestyle='dashed',label='threshold')
plt.show()

print(f'\nOverall Purity Score: {(float(total/len(X))*100):.4f}')
```
